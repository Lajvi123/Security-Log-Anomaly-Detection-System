```python
# Column names from the UNSW-NB15 feature list
columns = [
    'src_ip', 'src_port', 'dst_ip', 'dst_port', 'proto', 'service', 'state', 'dur', 'sbytes', 'dbytes',
    'sttl', 'dttl', 'sloss', 'dloss', 'service_type', 'Sload', 'Dload', 'Spkts', 'Dpkts',
    'swin', 'dwin', 'stcpb', 'dtcpb', 'smeansz', 'dmeansz', 'trans_depth', 'res_bdy_len', 'Sjit', 'Djit',
    'Stime', 'Ltime', 'Sintpkt', 'Dintpkt', 'tcprtt', 'synack', 'ackdat', 'is_sm_ips_ports', 'ct_state_ttl',
    'ct_flw_http_mthd', 'is_ftp_login', 'ct_ftp_cmd', 'ct_srv_src', 'ct_srv_dst', 'ct_dst_ltm', 'ct_src_ltm',
    'ct_src_dport_ltm', 'ct_dst_sport_ltm', 'ct_dst_src_ltm', 'attack_cat', 'label'
]

# Read the file and apply column names
data = pd.read_csv('/content/drive/MyDrive/UNSW-NB15_1.csv', names=columns, low_memory=False)
```

```python
#Covert object datatype into float
for col in data.columns:
    if data[col].dtype == 'object':
        try:
            data[col] = data[col].astype(float)
        except:
            pass
```

```python
#Encode categorical columns
label_encoders = {}
for col in data.select_dtypes(include='object').columns:
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col])
    label_encoders[col] = le
```

```python
#Normalize numeric features
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(data.drop(columns=['label']))
df_scaled = pd.DataFrame(X_scaled, columns=data.columns[:-1])
df_scaled['label'] = data['label'].values
```

```python
# Assign random labels: 90% normal (0), 10% attack (1) for realism
df_scaled['label'] = np.random.choice([0, 1], size=len(df_scaled), p=[0.9, 0.1])

# Save new simulated dataset
df_scaled.to_csv('/content/drive/MyDrive/preprocessed_unsw_nb15.csv', index=False)
print("Preprocessing complete. Data saved to: preprocessed_unsw_nb15.csv")
```
