```python
import pandas as pd
import time
import json
from datetime import datetime, timedelta
```

```python
file_path = '/content/drive/MyDrive/preprocessed_unsw_nb15.csv'
df = pd.read_csv(file_path)
```

```python
df.head(45)
```

```python
start_time = datetime.now()
df['timestamp'] = [start_time + timedelta(seconds=i) for i in range(len(df))]
```

```python
#convert a log into the a log
def generate_log(row):
    return {
        "timestamp": row['timestamp'].strftime("%Y-%m-%d %H:%M:%S"),
        "src_port": round(row.get('src_port', 0) * 65535),
        "dst_port": round(row.get('dst_port', 0) * 65535),
        "duration": round(row.get('dur', 0), 4),
        "sbytes": int(row.get('sbytes', 0)),
        "dbytes": int(row.get('dbytes', 0)),
        "label": int(row['label'])  # 0 = normal, 1 = attack
    }
```

```python
print("Starting log stream...\n")
for index, row in df.iterrows():
    log_entry = generate_log(row)
    print(json.dumps(log_entry, indent=2))
    time.sleep(0.3)  # simulate delay per log

    if index >= 30:
        break  # Stream first 30 logs for demo

print("\n Log stream complete.")
```
